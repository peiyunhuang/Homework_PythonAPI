```python
#Import Dependencies or Modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import time
from pprint import pprint

from citipy import citipy

#Import API Key
from config import api_key

output_data = "OutputData/Cities.csv"
```

```python
#Create Empty Lists to Hold Random Coordinates and Cities for the For Loop
latitudes_longitudes_list = []
cities_list = []

latitudes = np.random.uniform(low=-90.000, high=90.000, size=1500)
longitudes = np.random.uniform(low=-180.000, high=180.00, size=1500)
latitudes_longitudes_list = zip(latitudes, longitudes)

#Create a For Loop to Use Coordinates to Populate the List of Cities
for coordinates in latitudes_longitudes_list:
   city = citipy.nearest_city(coordinates[0], coordinates[1]).city_name
   city = city.replace(" ", "%20")
   if city not in cities_list:
       cities_list.append(city)
        
#Check the Length of the List of Cities
len(cities_list)
```

```python
#Create a Variable for the URL and API Key
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key
```

```python
#Create Empty Lists that Will Contain Information from the API Data
city_name = []
cloudiness = []
country = []
date = []
humidity = []
latitude = []
longitude = []
maximum_temperature = []
wind_speed = []

#Counter
record = 1
print(f"Commence Retrieval of Data")
print(f"-------------------------")

#Create a Nested For Loop with Try, Exception, and Continue
for city in cities_list:
   try:
       response = requests.get(f"{url}&q={city}").json()
       #pprint(response)
        
       city_name.append(response["name"])
       cloudiness.append(response["clouds"]["all"])
       country.append(response["sys"]["country"])
       date.append(response["dt"])
       humidity.append(response["main"]["humidity"])
       latitude.append(response["coord"]["lat"])
       longitude.append(response["coord"]["lon"])
       maximum_temperature.append(response["main"]["temp_max"])
       wind_speed.append(response["wind"]["speed"])
       city_record = response["name"]
       print(f"Processing Record {record} : {city_record}")
       print(f"{url}&q={city}")
       record += 1
       time.sleep(1.01)
   except:
       print("Unable to Locate City.  Skipping . . .")
   continue
```

```python
#Use the Form of a Dictionary to Create Headers and Index
weather_dictionary = {
   "City": city_name,
   "Cloudiness": cloudiness,
   "Country": country,
   "Date": date,
   "Humidity": humidity,
   "Lat": latitude,
   "Long": longitude,
   "Max Temperature": maximum_temperature,
   "Wind Speed": wind_speed
}
weather_data = pd.DataFrame(weather_dictionary)
weather_data.count()
```

```python
#Save DataFrame to CSV
weather_data.to_csv("WeatherData.csv")
weather_data.head()
```
